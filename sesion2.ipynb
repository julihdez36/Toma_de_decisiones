# Métodos numéricos

```python
# Medidas de centralidad: media

print('Media de ventas:', np.mean(forbes21['Sales']))

# Medidas de centralidad: mediana

print('Mediana de ventas:', np.median(forbes21['Sales']))
```

```python
forbes21['Sales'].mean(), forbes21['Sales'].median()
```

```python
edades = [25,25,26,28,30,75]
edades = pd.Series(edades)
```

```python
print(edades.mean()) #promedio de edades; muy sensible a los outliers; mean and average
print(edades.median()) #mediana de edades
```

```python
# Medidas de dispersión

print(edades.var()) # una medida de dispersión cuya escala es cuadratica
print(edades.std()) # Es la raíz de la varianza
```

```python
# Coeficiente de variación std()/mean()

print('El coeficiente de variación es ',100 * edades.std()/edades.mean(),'%') # CV < 15% : data homogenea
```

```python
# Medidas de posición
# Percentiles
edades = [25,25,26,28,28,30,30]
edades = pd.Series(edades)
edades
```

```python
edades.median() # Mediana = Percentil 50 = al cuartil 2
edades.quantile([0.25,0.5,0.75]) # Cuartiles
```

## Gráficos para variables cualitativas

```python
# Bar plot

df_suda = forbes21[forbes21['Country'].isin(suda)]
```

```python
freq_suda = df_suda['Country'].value_counts(normalize= True) *100
```

```python
#plt.bar(freq_suda )
plt.bar(freq_suda.index, freq_suda.values,color = ['purple','red','blue','orange','g','y'])
plt.title('Diagrama de barras %')
plt.ylabel('Porcentaje de participación')
plt.xlabel('Paises')
plt.grid(alpha=.4)
plt.show()
```

```python
# Gráficos de sectores
# Gráfico de torta

plt.pie(freq_suda.values, labels= freq_suda.index,shadow= True)
plt.title('Empresas sudamericanas por paises')
plt.show()
```
